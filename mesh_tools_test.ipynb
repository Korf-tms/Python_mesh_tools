```python
import numpy as np
import mesh_tools
```

```python
# rectangle (aX,bX) x (aY,bY) with discretization nX x nY:
aX = 0
bX = 2
nX = 3
aY = 0
bY = 1
nY = 3

# coordinates of nodes:
coord_X = np.linspace(aX, bX, nX)
coord_Y = np.linspace(aY, bY, nY)
x, y = np.meshgrid(coord_X, coord_Y)
rectangle_node_X = x.flatten()
rectangle_node_Y = y.flatten()

# triangulation:
rectangle_elem = mesh_tools.rectangle_triangulation(nX, nY)
```

```python
# test of some implemented tools for 2d mesh postprocessing:
mesh = mesh_tools.Mesh2d(rectangle_node_X, rectangle_node_Y, rectangle_elem)
mesh.plot_mesh()
mesh.plot_elem_centers(show_numbers=True)
mesh.plot_edge_midpoints()
mesh.plot_nodes(show_numbers=True)
print("Elements (in rows):")
print(mesh.elem.T)  # indices nodes for all (triangular) elements
print("Edges (in rows):")
print(mesh.edge.T)  # indices of nodes for all edges
print("Boundary flags:")
print(mesh.edge_boundary_flag)  # boundary flags for all edges
```

```python
mesh.P1_to_PN(N=4)

mesh.plot_mesh()
mesh.plot_nodes(show_numbers=True)
mesh.plot_elem_centers(show_numbers=True)

print("Elements including added points (in rows):")
print(mesh.elem_PN.T)  # indices nodes for all elements including added points
```
